```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
```

```python
data=pd.read_excel('dataset/Fasle_Bahar_94_Pasargad.xls')
```

```python
data
```

```python
Y=data['Sum']
X=data[['Sesion','Month','DayOfNumber','DayOfWeek','OnOffDay','Hour']]
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3)
```

```python
x3=['identity','logistic','tanh','relu']
Max=0
```

```python
for X1 in range(25,50,1):
    for X2 in np.arange(0.1,0.2,0.1):
        for X3 in range(0,4,1):
            for X4 in range (100,200,10):
                model=MLPRegressor(hidden_layer_sizes=(X1),learning_rate_init=X2,activation=x3[X3],max_iter=X4)
                model.fit(Xtrain,Ytrain)
                pred=model.predict(Xtest)
                acc=r2_score(Ytest,pred)
                if acc>Max:
                    Max=acc
                    w=[X1,X2,X3,X4]
                   
```

```python
print ("r2_score is",Max)
print (w)
```
